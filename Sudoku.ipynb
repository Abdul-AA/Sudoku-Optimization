```python
import gurobipy as gb
from gurobipy import *
```

```python
# A list of strings from "1" to "9" is created
Sequence = ["1", "2", "3", "4", "5", "6", "7", "8", "9"]

Rows = Sequence
Cols = Sequence
Vals = Sequence
```

```python
# The boxes list is created, with the row and column index of each square in each box
Boxes =[]
for i in range(3):
    for j in range(3):
        Boxes += [[(Rows[3*i+k],Cols[3*j+l]) for k in range(3) for l in range(3)]]
```

```python
prob=gb.Model("Sudoku")
```

```python
X=prob.addVars(Rows, Cols, Vals, vtype=GRB.BINARY, name=[f'Row {r}, Col {c}, Val {v}' for r in Rows for c in Cols for v in Vals])
```

```python
prob.setObjective(0, GRB.MINIMIZE)
```

```python
prob.addConstrs((quicksum(X[i,j,k] for k in Vals) == 1 for i in Rows for j in Cols))
```

```python
for k in Vals:
    for box in Boxes:
        prob.addConstr((quicksum(X[i,j,k] for (i,j) in box) == 1))
        
```

```python
prob.addConstrs((quicksum(X[i,j,k] for j in Cols) == 1 for i in Rows for k in Vals))
```

```python
prob.addConstrs((quicksum(X[i,j,k] for i in Rows) == 1 for j in Cols for k in Vals))
```

```python
game=[[0,0,3,0,0,0,0,9,0],
      [0,0,0,0,0,9,5,0,0],
      [0,0,0,8,4,3,0,0,0],
      [0,0,0,0,0,0,7,0,2],
      [0,0,0,0,0,6,0,0,0],
      [9,0,2,0,3,0,0,0,0],
      [0,0,0,0,0,8,0,0,0],
      [6,0,5,0,0,0,9,0,7],
      [1,0,7,0,0,0,2,5,0]
]
```

```python
for i in range(9):
    for j in range(9):
        if game[i][j] != 0:
            prob.addConstr(X[Rows[i], Cols[j], str(game[i][j])] == 1)
```

```python
prob.optimize()
```

```python
for v in prob.getVars():
    if v.x > 0.01:
        print(v.varName, v.x)
```

```python
solution = [[0]*9 for _ in range(9)]
for i in Rows:
    for j in Cols:
        for v in Vals:
            if X[i,j,v].x > 0.5:
                solution[int(i)-1][int(j)-1] = int(v)

# Print the solution
for row in solution:
    print(row)
```
